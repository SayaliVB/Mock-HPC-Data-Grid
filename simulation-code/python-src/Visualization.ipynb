```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assuming a 4-core system for the example
cpu_cores = 8

# Number of processes and their respective execution times
process_counts = np.array([4, 6, 8, 10, 12])
execution_time = np.array([
    np.mean([7.51644, 7.52026, 7.52418]),  # for 4 processes
    np.mean([5.86601, 5.8735, 5.87652, 5.90471, 5.91064]),
    np.mean([5.54843, 5.5546, 5.56185, 5.58705, 5.5917, 5.59559, 5.60578]),  # for 8 processes
    np.mean([6.04654, 6.07829, 6.0785, 6.08263, 6.08449, 6.09528, 6.09828, 6.10025, 6.10617]),
    np.mean([5.8419, 5.83883, 5.85138, 5.8778, 5.88365, 5.89181, 5.89233, 5.90257, 5.93471, 5.93208, 5.9521])  # for 12 processes
])

# Create the bar chart
fig, ax = plt.subplots()
bars = ax.bar(process_counts, execution_time, color='orange')

# Labels and title
ax.set_xlabel('Number of Processes')
ax.set_ylabel('Average Execution Time (seconds)')
ax.set_title('Average Execution Time by Number of Processes(MPI)')

# Annotate bars with over-subscribing and under-subscribing
for idx, bar in enumerate(bars):
    height = bar.get_height()
    if process_counts[idx] < cpu_cores:
        label = 'Under-subscribing'
    elif process_counts[idx] > cpu_cores:
        label = 'Over-subscribing'
    else:
        label = 'Balanced-subscribing'

    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{label}',
            ha='center', va='bottom')

plt.xticks(process_counts, [f'{pc} Processes' for pc in process_counts])

# Show the plot
plt.show()
```

```python

# Number of threads and their respective execution times
thread_counts = np.array([4, 6, 8, 10, 12])
execution_time = np.array([
    np.mean([6.02757, 6.03256, 6.04128, 6.04667]),  # for 4 threads
    np.mean([3.91031, 3.93271, 3.95044, 3.95398, 3.95488, 4.0408]),
    np.mean([3.84011, 3.89542, 3.93596, 3.97141, 3.9813, 3.98491, 4.00263, 4.00698]),  # for 8 threads
    np.mean([5.57698, 5.58545, 5.61635, 5.62852, 5.65653, 5.65769, 5.66648, 5.67593, 5.72882, 5.74054]),
    np.mean([5.62301, 5.63629, 5.6441, 5.648, 5.67516, 5.67563, 5.71135, 5.71396, 5.71761, 5.72024, 5.72772, 5.75422])  # for 12 threads
])

# Create the bar chart
fig, ax = plt.subplots()
bars = ax.bar(thread_counts, execution_time, color='orange')

# Labels and title
ax.set_xlabel('Number of threads')
ax.set_ylabel('Average Execution Time (seconds)')
ax.set_title('Average Execution Time by Number of threads(OpenMP)')

plt.xticks(thread_counts, [f'{pc} threads' for pc in thread_counts])

# Show the plot
plt.show()
```
